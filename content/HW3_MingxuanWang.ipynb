```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
%matplotlib inline
```

```python
malaria_deaths_age=pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2018/2018-11-13/malaria_deaths_age.csv")
malaria_deaths_age
```

# Plot 1: Deaths over the Year of Each Country

```python
malaria_deaths_age_groupby_code = malaria_deaths_age.groupby(['code','year'],as_index=False).sum()

year=widgets.Play(
    value=1990,
    min=1990,
    max=2016,
    step=1,
    interval=500,
    description='Year:',
    disabled=False
)

slider = widgets.IntSlider(
    value=1990,
    min=1990,
    max=2016,
    step=1,
    description='Year:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
widgets.jslink((year, 'value'), (slider, 'value'))


def f(year):
    fig = px.choropleth(malaria_deaths_age_groupby_code[malaria_deaths_age_groupby_code['year']==year], locations='code', color='deaths',
                               color_continuous_scale="reds",
                               range_color=(0, 50000),
                               labels={'deaths':'deaths'}
                              )
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    fig.show()
    
out = widgets.interactive_output(f, {'year': year})

widgets.VBox([widgets.HBox([year, slider]), out])



```

# Plot 2: Deaths over the Years for Each Country in Each Age Group

```python
entity = widgets.Dropdown(
    options=set(malaria_deaths_age['entity']),
    value='Ethiopia',
    description='Entity:',
    disabled=False,
)

def f(entity):
    fig = px.line(malaria_deaths_age[malaria_deaths_age['entity']==entity], x="year", y="deaths",color="age_group")
    fig.show()

out = widgets.interactive_output(f, {'entity': entity})

widgets.VBox([widgets.VBox([entity]), out])
```

# Plot 3: Pie Chart of Deaths for Age Groups of Each Country

```python
malaria_deaths_age_groupby_entity = malaria_deaths_age.groupby(['entity','age_group'],as_index=False).sum()

entity = widgets.Dropdown(
    options=set(malaria_deaths_age_groupby_entity['entity']),
    value='Ethiopia',
    description='Entity:',
    disabled=False,
)

def f(entity):
    fig = px.pie(malaria_deaths_age_groupby_entity[malaria_deaths_age_groupby_entity['entity']==entity], values="deaths", names="age_group",color_discrete_sequence=px.colors.qualitative.Set3,title="A")
    fig.show()

    
out = widgets.interactive_output(f, {'entity': entity})

widgets.VBox([widgets.VBox([entity]), out])
```
