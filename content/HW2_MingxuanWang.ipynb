## import packages

```python
import numpy as np
import sympy as sym
```

## function to generate an arbitrary large expansion

```python
def pi_expan(num,n):
    """
    Parameters
    ----------
    num : int
        Number that needs to be generated into a large expansion.
    n : int
        Number of digits follows the decimal point and needs to be included in the expansion.
        
    Returns
    -------
    str_pi : str
         The expansion of the number.
    """
    str_pi = str(int(num.evalf(n)*10**n))
    return str_pi
```

## function to check if a number is prime

```python
def sliding_window(num,k):
    """
    Parameters
    ----------
    num : int
        Number that needs to be used to generate sliding windows.
    k : int
        Length of the sliding windows.
        
    Returns
    -------
    windows : list
         List of sliding windows.
    """
    windows=[]
    for i in range(len(num)-k+1):
        windows.append(int(num[i:k+i]))
    return windows
```

## function to generate sliding windows

```python
def is_prime(n):
    """
    Parameters
    ----------
    n : int
        Number that need to be judged to be prime.
    lb : list
        list of the label signify represent the number under the square root of n is prime or not.
    primes : ndarray
        array of the primes under the square root of n.
        
    Returns
    -------
    res : bool
         Whether the number is a prime or not.
    """
    sqrt_n=int(np.sqrt(n))
    lb=[True]*(sqrt_n+1)
    lb[0]=False
    lb[1]=False
    for i in range(2,int(np.sqrt(sqrt_n)+1)):
        if lb[i] == True:
            for j in range(i**2, sqrt_n+1, i):
                lb[j]=False
    primes=np.nonzero(lb)[0]
    flag = True
    for prime in primes:
        if n % prime == 0:
            flag = False
            break
    if flag:
        res=True
        return True
    else:
        res=False
        return False
```

## final function

```python
def find_prime(num,n,k):
    """
    Parameters
    ----------
    num : int
        Number that needs to be used to find the first k digits prime.
    n : int
        Number of digits follows the decimal point and needs to be included in the expansion.
    k : int
        Length of the sliding windows.
        
    Returns
    -------
    window : int
        First k digits prime finded.
    """
    for window in sliding_window(pi_expan(num,n),k):
        if is_prime(window):
            return window
```
